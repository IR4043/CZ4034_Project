## Remove Non-English Reviews

```python
from langdetect import detect

# Function to detect language of a tweet
def detect_language(text):
    try:
        lang = detect(text)
        return lang
    except:
        return 'unknown'

# Filter out non-English tweets
df['language'] = df['reviewDescription'].apply(detect_language)
df = df[df['language'] == 'en']

# Drop the 'language' column
df = df.drop(columns=['language'])
```

```python
df
```

```python
# Select the columns you want to keep
columns_to_keep = ['title', 'reviewDescription']

# Keep only the selected columns
df_process = df.loc[:, columns_to_keep]
```

```python
df_process
```

```python
import re

# Define a regular expression pattern to match the weird characters
pattern = re.compile("[^\x00-\x7F]+")

# Replace the weird characters with an empty string
df['title'] = df['title'].apply(lambda x: pattern.sub('', x))
df['reviewDescription'] = df['reviewDescription'].apply(lambda x: pattern.sub('', x))
```
